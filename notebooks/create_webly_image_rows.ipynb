```python
# Append on the items trained after for the faster comparison
for p in thesis_root.iterdir():
    filename = p.stem
    if filename.endswith('-ranking'):
        with open(p) as f:
            ranking = json.load(f)
        obj = filename[:-(len('-ranking'))]
        webly_object_ranking[obj] = ranking

print(webly_object_ranking)
    
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import gridspec

    
def row(img_paths, f_name):
#     plt.figure(figsize=(15,5))
#     plt.subplots_adjust(bottom=0.0, top=0.5, hspace=0.1)

    nrow = 2
    ncol = 5

    fig = plt.figure(figsize=(ncol+1, nrow+1)) 

    gs = gridspec.GridSpec(nrow, ncol,
             wspace=0.1, hspace=0.05,
             top=1., bottom=0.0, 
             left=0.0, right=1.0) 

    for i in range(nrow):
        for j in range(ncol):
            im = mpimg.imread(img_paths[i * ncol + j])
            ax = plt.subplot(gs[i,j])
            ax.imshow(im)
            ax.axis("off")

    plt.savefig(f_name)
```

```python
import random
for obj, search in target_objects.items():
    search = search.replace(' ', '_')
    obj_root = Path(crop_root / obj)
    
    webly_paths = list(map(lambda x: Path(webly_root / x), webly_object_ranking[obj][:5]))
    
    arid_crops = []
    for p in obj_root.iterdir():
        arid_crops.append(p)
    arid_crops = random.choices(arid_crops, k=5)
    webly_paths.extend(arid_crops)
    row(webly_paths, f'/home/justin/Desktop/thesis/obj_imgs/{obj}.png')
                                  
    
```
